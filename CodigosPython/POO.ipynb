```python
#Slider

import numpy as np
import matplotlib.pyplot as plt 
from matplotlib.widgets import Slider, Button, RadioButtons

fig, ax = plt.subplots()
plt.subplots_adjust(left=0.25, bottom=0.25)
t = np.arange(0.0, 1.0, 0.001)
a = 5
f = 3
s = a*np.sin(2*np.pi*f*t)
l, = plt.plot(t, s, lw=2, color="red")
plt.axis([0, 1, -10, 10])

axcolor = "blue"
axfreq = plt.axes([0.25, 0.1, 0.65, 0.03], facecolor=axcolor)
axamp = plt.axes([0.25, 0.15, 0.65, 0.03], facecolor=axcolor)

sfreq = Slider(axfreq, "Frecuencia", 0.1, 30.0, valinit = f)
samp = Slider(axamp, "Amplitud", 0.1, 10.1, valinit = a)

def update(val):
    amp = samp.val
    freq = sfreq.val
    l.set_ydata(amp*np.sin(2*np.pi*freq*t))
    fig.canvas.draw.idle()
    
sfreq.on_changed(update)
samp.on_changed(update)

resetax = plt.axes([0.8, 0.025, 0.1, 0.04])
button = Button(resetax, "Reset", color=axcolor)

def reset(event):
    sfreq.reset()
    samp.reset()
    
button.on_clicked(reset)

rax = plt.axes([0.025, 0.5, 0.15, 0.15], facecolor=axcolor)
radio = RadioButtons(rax, ("blue", "green", "red"), active=0)

def colorfunc(label):
    l.set_color(label)
    fig.canvas.draw_idle()
    
radio.on_clicked(colorfunc)

plt.show()
```

```python
#Scopes #Si puedo evitar usar variables locales, lo hare

def scope_test():
    def do_local():
        spam = "local spam"
        
    def do_nonlocal():
        nonlocal spam
        spam = "nonlocal spam"
        
    def do_global():
        global spam
        spam = "global spam"
        
    spam = "test spam"
    
    do_local()
    print("After local assigment: ", spam)
    
    do_nonlocal()
    print("After nonlocal assigment: ", spam)
    
    do_global()
    print("After nonlocal assigment: ", spam)
    
scope_test()
print("In global scope: ", spam)
```

```python
#Llamar a traves de la instantia 


class Complex:
    def __init__(self, r=4, i=7):
        self.r = r
        self.i = i
        
    def mostrar(self):
        print(str(self.r)+ " + i"+str(self.i))
        
```

```python
#Atributos de datos: Podemos crear atributos dentro de una instancia. 
#Pero solo la instancia!!! no se modifican los atributos de las clases

#Atributos de instancia (se definen mediante self: referencia a la instancia)
#Atributos de clase(me van a mandar al psiquiatrico: instancias generadas por una clase)

#Se prioriza el atributo de instancia

class dog:
    
    familia = "canine" #atributo de clase
    def __init__(self, name):
        self.name = name # atributo de instancia

        d = dog("Fido")
    #d = referencia
    #instanciar de es d=dog("fido")
        e = dog("Buddy")
    
        d.familia
        "canine"
        e.familia
        "canine"
    
        d.name
        "Fido"
        e.name
        "Buddy"
```

```python
#Metodos de instancia: Metodo: funcion que pertenece a determinada clase y se puede llamar a traves de la instancia 
#Clase para instanciar perros 

class dog:
    
    def __init__(self, name):
        self.name = name # atributo de instancia
        self.tricks = [] #lista vacia para cada dog
        
    def aprende(self, trick):
        #self.tricks.append(trick)
        self.tricks = trick
        
        d = dog("yetro")
        e = dog("isra")
    
        d.aprende("sientate")
        e.aprende("da la pata")
    

        d.trick
        ["sentarse"]
        e.trick
        ["da la pata"]
```
